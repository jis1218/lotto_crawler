```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pymysql
import time
```

```python
host = "localhost"
port = 3309
database = "insup"
username = "root"
password = "amadeus21!"
```

```python
conn = pymysql.connect(
    host=host,
    user=username,
    password=password,
    database=database,
    port=port
)
```

```python
for drawNo in range(262, 1066):
    response = urlopen(f"https://www.dhlottery.co.kr/store.do?method=topStore&pageGubun=L645&nowPage=1&drwNo={drawNo}")
    soup = BeautifulSoup(response, "html.parser")
    total_page = len(soup.find("div", {"class":"paginate_common"}).find_all("a"))
    for nowPage in range(1, total_page+1):
        response = urlopen(f"https://www.dhlottery.co.kr/store.do?method=topStore&pageGubun=L645&nowPage={nowPage}&drwNo={drawNo}")
        soup = BeautifulSoup(response, "html.parser")
        table = soup.find_all("table", {"class":"tbl_data tbl_data_col"})
        tbody = table[1].find("tbody")
        trs = tbody.find_all("tr")
        if trs[0].find('td').has_attr('class'):
            break
        for tr in trs:
            cursor = conn.cursor()
            td = tr.find_all("td")
            

            store_name = td[1].get_text().replace("'", "").strip()
            address = td[2].get_text().strip()
            query = f'INSERT INTO second_lottery_store(round, store_name, address) VALUES({drawNo}, \'{store_name}\', \'{address}\')'

            cursor.execute(query)

            conn.commit()

            cursor.close()

        time.sleep(2)
    print(str(drawNo), '회차 완료')
        
        
```
